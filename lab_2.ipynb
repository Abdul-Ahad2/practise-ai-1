```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Task 1

arr1D = np.array([x for x in range(0,10)])

arr2D = np.array([x for x in range(1,10)]).reshape(3,3)

arr3D = np.random.randint(1, 100, size=(2, 3, 4))

print(arr1D)
print(arr2D)
print(arr3D)
```

```python
# Task 2

print(arr2D[1])

print(arr2D[2][1])

print(arr3D[0])
```

```python
# Task 3

reshapedArr = arr1D.reshape(2,5)
print(reshapedArr)
```

```python
# Task 4

print(arr1D + 10)

print(arr2D * 2)
```

```python
# Task 1: Basic Line Plot

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot")
plt.grid(True)
plt.show()
```

```python
# Task 2: Scatter Plot

x2 = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y2 = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x2, y2, c='r', marker='x')
plt.show()
```

```python
# Task 3: Bar Plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.bar(categories, values, color=colors)
plt.show()
```

```python
# Task 4: Histogram

nums = np.random.randn(1000)
plt.hist(nums, 30)
plt.show()
```

```python
# Task 5: Pie Chart

label = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]
explodes = [0.1, 0.0, 0.0, 0.0]

plt.pie(sizes, explodes, label)
plt.show()
```
